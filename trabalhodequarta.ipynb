```python
#Vetor com 1.000 números aleatórios
import random
import time

def bubble_sort(vetor):
    x = len(vetor)
    for i in range(x):
        for y in range(x - 1):
            if vetor[y] > vetor[y + 1]:
                aux = vetor[y]
                vetor[y] = vetor[y + 1]
                vetor[y + 1] = aux
    return vetor

def merge(vetor, left, middle, right):
    x1 = middle - left + 1
    x2 = right - middle
    Left = vetor[left:middle + 1]
    Right = vetor[middle + 1:right + 1]

    a = 0
    b = 0
    c = left
    while a < x1 and b < x2:
        if Left[a] <= Right[b]:
            vetor[c] = Left[a]
            a += 1
        else:
            vetor[c] = Right[b]
            b += 1
        c += 1
    while a < x1:
        vetor[c] = Left[a]
        a += 1
        c += 1
    while b < x2:
        vetor[c] = Right[b]
        b += 1
        c += 1

def merge_sort(vetor, left, right):
    if left < right:
        middle = left + (right - left) // 2
        merge_sort(vetor, left, middle)
        merge_sort(vetor, middle + 1, right)
        merge(vetor, left, middle, right)

def main():
    tam_vetores = 1000


    A = [random.randint(0, 99) for _ in range(tam_vetores)]

    start = time.time()
    bubble_sort(A)
    end = time.time()
    diff_bubble = end - start

    B = [random.randint(0, 99) for _ in range(tam_vetores)]

    start = time.time()
    merge_sort(B, 0, tam_vetores - 1)
    end = time.time()

    diff_merge = end - start

    print("\nPara um vetor aleatório de tamanho", tam_vetores)
    print(f"O tempo de execução do Bubble Sort será de : {diff_bubble:.3f} segundos")
    print(f"O tempo de execução do Merge Sort será de : {diff_merge:.3f} segundos")

if __name__ == "__main__":
    main()
```
